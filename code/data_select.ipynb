```python
import pandas as pd
```

```python
data = pd.read_csv('raw_data.csv', encoding='cp949')
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data = data.drop(['code','분류명'], axis=1)
```

```python
data
```

```python
data_count = data.groupby('c_name')['c_index'].count().sort_values(ascending=False)
```

```python
data_count
```

```python
list_name = ['정보통신업', '건설업', '도소매업', '음식점', '기기류 제조업', '전문과학기술업']

new_data = data[(data['c_name']=='정보통신업')|(data['c_name']=='건설업')|(data['c_name']=='도소매업')|(data['c_name']=='음식점')|(data['c_name']=='기기류 제조업')|(data['c_name']=='전문과학기술업')]
```

```python
new_data = new_data.sort_values('c_index').reset_index()
```
