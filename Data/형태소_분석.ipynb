```python
import pandas as pd
import numpy as np
import csv

import os
import re
from PIL import Image
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt

crawling_data = pd.read_csv('/content/글로벌에이스.csv',encoding ='utf-8' )
```

```python
texts = crawling_data.get("기사내용")
print(texts)
```

```python
print(type(texts))
```

```python
text_list=[]
for i in texts:
  text_list.append(i)

print(text_list)
print(type(text_list))
```

```python
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
morph_list=[]
for i in text_list:
  morphs = twitter.nouns(i)
  for morph in morphs:
    morph_list.append(morph)
```

```python
print(morph_list)
```

```python
len(morph_list)
```

```python
from collections import Counter
count = Counter(morph_list)
```

```python
# 갯수확인
count.most_common()
```

```python
# 불용어처리 된 데이터 텍스트로 저장
for i in morph_list:
  f = open("morph.txt","a",encoding="utf8")
  f.write(i+'\n')
  f.close()
```
